```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
```

```python
N_SAMPLES = 20
N_FEAT = 2

x = torch.linspace(-3.14, 3.14, N_SAMPLES)
#x = torch.linspace(0, 10, N_SAMPLES, dtype=int)
xgrids = torch.meshgrid((x,) * N_FEAT, indexing='ij')
X = torch.stack([g.reshape(-1) for g in xgrids], dim=1)

y = torch.sin(X).sum(dim=1)
```

```python
plt.figure(figsize=(8, 6))
plt.imshow(y.reshape(N_SAMPLES, N_SAMPLES),
           extent=[x.min(), x.max(), x.min(), x.max()],
           origin='lower', cmap='viridis')
plt.colorbar(label='y')
plt.xlabel('x1')
plt.ylabel('x2')
#plt.show()
```
